```python
df2=df1.drop(['Accident Date','Day_of_Week','Junction_Control','Junction_Detail','Local_Authority_(District)','Number_of_Vehicles','Police_Force','Road_Surface_Conditions'
             ,'Road_Type','Time','Weather_Conditions','Month','Day'],axis=1)
df2.head(2)
```

```python
x=df2.drop('Accident_Severity',axis=1)
y=df2['Accident_Severity']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
X=sc.fit_transform(x)
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
m=[]
m.append(('lo',LogisticRegression()))
m.append(('dt',DecisionTreeClassifier()))
m.append(('rf',RandomForestClassifier()))
m.append(('Kn',KNeighborsClassifier()))
```

```python
m
```

```python
from sklearn.model_selection import KFold,cross_val_score
```

```python
for i ,l in m:
    k=KFold(n_splits=10)
    c=cross_val_score(l,x_train,y_train,cv=k)
    msg='%s:%f (%f)' %(i,c.mean(),c.std())
    print(msg)
```

```python
lo=LogisticRegression()
lo.fit(x_train,y_train)
```

```python
lo_pred=lo.predict(x_test)
```

```python
lo_pred
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,lo_pred))
```

```python
from imblearn.over_sampling import SMOTE
```

```python
x_resamp,y_resamp=SMOTE().fit_resample(X,y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x_resamp,y_resamp,test_size=0.25,random_state=0,stratify=y_resamp)
```

```python
for i ,l in m:
    k=KFold(n_splits=10)
    c1=cross_val_score(l,X_train,Y_train,cv=k)
    msg='%s:%f (%f)' %(i,c1.mean(),c1.std())
    print(msg)
```

```python
lo1=LogisticRegression()
lo1.fit(X_train,Y_train)
```

```python
y_pred=lo1.predict(X_test)
y_pred
```

```python
print(classification_report(y_pred,Y_test))
```
