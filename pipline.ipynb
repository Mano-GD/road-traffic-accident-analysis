```python
import pandas as pd
```

```python
df=pd.read_csv('clean_dt_acc.csv')
df.head(1)
```

```python
df1=df.iloc[:,1:]
df1.head(2)
```

```python
df2=df1.drop(['Accident Date','Day_of_Week','Junction_Control','Junction_Detail','Local_Authority_(District)','Number_of_Vehicles','Police_Force','Road_Surface_Conditions'
             ,'Road_Type','Time','Weather_Conditions','Month','Day'],axis=1)
df2.head(2)
```

```python
obj=df2.select_dtypes('object').columns
```

```python
df2.describe()
```

```python
for l in obj:
    v=df2[l].value_counts()
    print(v)
```

```python
x=df2.drop('Accident_Severity',axis=1)
y=df2['Accident_Severity']
```

```python
from sklearn_pandas import DataFrameMapper
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.ensemble import RandomForestClassifier
```

```python
dm=DataFrameMapper([(['Light_Conditions',LabelEncoder()]),(['Urban_or_Rural_Area',LabelEncoder()]),(['Vehicle_Type',LabelEncoder()]),
                    (['Year',LabelEncoder()])])
dm
```

```python
m=[]
m.append(['Mapper',dm])
m.append(['Sc',StandardScaler()])
m.append(['Rf',RandomForestClassifier()])
m
```

```python
from sklearn.pipeline import Pipeline
```

```python
m1=Pipeline(m)
m1
```

```python
m1.fit(x,y)
```
